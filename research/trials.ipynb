```python
#ensure annotations
def get_prodict(x:int,y:int)->int:
    return x*y
```

```python
get_prodict(x=2,y=3)
```

```python
get_prodict(x=2,y='3')
```

```python
from ensure import ensure_annotations
```

```python
get_prodict(x=2,y='3')
```

```python
@ensure_annotations
def get_prodict(x:int,y:int)->int:
    return x*y
```

```python
get_prodict(x=2,y='3')
```
